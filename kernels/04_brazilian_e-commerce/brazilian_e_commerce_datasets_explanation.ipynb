## Testes Join

Temos os seguintes datasets:
    
    df_customers
    df_geo
    df_order_items
    df_order_payments
    df_order_reviews
    df_orders
    df_prod
    df_sellers    

```python
# Unindo tudo o que é possível
print(f'Dimensões de df_orders: {df_orders.shape}')
print(f'Dimensões de df_order_payments: {df_order_payments.shape}')
df_orders_and_payments = df_orders.merge(df_order_payments, on='order_id', how='inner')

print(f'Dimensão da união: {df_orders_and_payments.shape}')

# Primeiras linhas
df_orders_and_payments.head()
```

```python
# Verificando
df_orders_and_payments.isnull().any()
```

```python
# Verificando entradas duplicadas
df_orders_and_payments[df_orders_and_payments.duplicated('order_id', keep=False) == True]
```

order_ids duplicados indicam diferentes meios de pagamentos escolhidos por clientes para uma mesmo compra (ex, credit card, voucher, ...). As diferenças encontram-se nas colunas `payment_sequential` e `payment_type`.

```python
# Unindo customers
print(f'Dimensões de df_orders: {df_orders.shape}')
print(f'Dimensões de df_order_payments: {df_order_payments.shape}')
print(f'Dimensões de df_customers: {df_customers.shape}')

df_ord_pay_cust = df_orders_and_payments.merge(df_customers, on='customer_id', how='inner')
print(f'\nDimensão da união orders and payments: {df_orders_and_payments.shape}')
print(f'Dimensão da união ord pay and costumers: {df_ord_pay_cust.shape}')

# Primeiras linhas
df_ord_pay_cust.head()
```

Para análises futuras, não é interessante manter customers dentro deste dataset, visto que, como os pagamentos diferentes geram ids repetidos, contabilizações erradas podem ser realizadas, por exemplo, no levantamento de quantas compra foram realizadas em SP (a nao ser que seja realizado um Group By).

```python
# Juntando review
print(f'Dimensões de df_orders: {df_orders.shape}')
print(f'Dimensões de df_order_payments: {df_order_payments.shape}')
print(f'Dimensões de df_customers: {df_customers.shape}')
print(f'Dimensões de df_reviews: {df_order_reviews.shape}')

df_ord_pay_cust_rev = df_ord_pay_cust.merge(df_order_reviews, on='order_id', how='inner')
print(f'\nDimensão da união orders and payments: {df_orders_and_payments.shape}')
print(f'Dimensão da união ord pay and costumers: {df_ord_pay_cust.shape}')
print(f'Dimensão da união ord pay cust review: {df_ord_pay_cust_rev.shape}')

# Primeiras linhas
df_ord_pay_cust_rev.head()
```

```python
# Unindo itens
# Juntando review
print(f'Dimensões de df_orders: {df_orders.shape}')
print(f'Dimensões de df_order_payments: {df_order_payments.shape}')
print(f'Dimensões de df_customers: {df_customers.shape}')
print(f'Dimensões de df_reviews: {df_order_reviews.shape}')
print(f'Dimensões de df_order_itens: {df_order_items.shape}')

df_all = df_ord_pay_cust_rev.merge(df_order_items, on='order_id', how='inner')
print(f'\nDimensão da união orders and payments: {df_orders_and_payments.shape}')
print(f'Dimensão da união ord pay and costumers: {df_ord_pay_cust.shape}')
print(f'Dimensão da união ord pay cust review: {df_ord_pay_cust_rev.shape}')
print(f'Dimensão de tudo unido: {df_all.shape}')

# Primeiras linhas
df_all.head()
```

```python
# Unindo sellers
print(f'Dimensões de df_orders: {df_orders.shape}')
print(f'Dimensões de df_order_payments: {df_order_payments.shape}')
print(f'Dimensões de df_customers: {df_customers.shape}')
print(f'Dimensões de df_reviews: {df_order_reviews.shape}')
print(f'Dimensões de df_order_itens: {df_order_items.shape}')
print(f'Dimensão de df_sellers: {df_sellers.shape}')

df_all = df_all.merge(df_sellers, on='seller_id', how='inner')
print(f'\nDimensão da união orders and payments: {df_orders_and_payments.shape}')
print(f'Dimensão da união ord pay and costumers: {df_ord_pay_cust.shape}')
print(f'Dimensão da união ord pay cust review: {df_ord_pay_cust_rev.shape}')
print(f'Dimensão de tudo unido: {df_all.shape}')

# Primeiras linhas
df_all.head()
```
