Let's define some useful functions to be used along this kernel.

```python
def format_spines(ax, right_border=True):
    """
    This is a usefull function used for make plots more beautiful
    
    Input:
        ax: matplotlib axis
        right_border: boolean flag for deciding or not to plot the right border of axis
    """    
    ax.spines['bottom'].set_color('#CCCCCC')
    ax.spines['left'].set_color('#CCCCCC')
    ax.spines['top'].set_visible(False)
    if right_border:
        ax.spines['right'].set_color('#CCCCCC')
    else:
        ax.spines['right'].set_color('#FFFFFF')
    ax.patch.set_facecolor('#FFFFFF')
    
def donut_plot(col, ax, df, text='', colors=['navy', 'crimson'], labels=['non-fraud', 'fraud']):
    """
    This function plots a customized donut chart of the target column (binary task)
    
    Input:
        col: target column of a binary classification task
        ax: matplotlib axis
        df: DataFrame object with the data
        text: the text to be plotted on the center of the donut chart
        colors: list of two colors used to identify the target class
        labels: list of labels to describe the target class
    Output:
        a customized donut chart
    """
    sizes = df[col].value_counts().values
    center_circle = plt.Circle((0,0), 0.80, color='white')
    ax.pie((sizes[0], sizes[1]), labels=labels, colors=colors, autopct='%1.2f%%')
    ax.add_artist(center_circle)
    kwargs = dict(size=20, fontweight='bold', va='center')
    ax.text(0, 0, text, ha='center', **kwargs)

def plot_confusion_matrix(cm, classes, title='Confusion matrix', cmap=plt.cm.Blues):
    """
    This function sets up and plot a really cool Confusion Matrix
    
    Input:
        cm: confusion matrix generated by sklearn's function confusion_matrix(set, predictions)
        classes: target labels to be plotted on confusion matrix
        title: plot title
        cmap: colors of confusion matrix
    Output:
        A beautiful confusion matrix
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, fontsize=14)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    # Plot configuration
    thresh = cm.max() / 1.2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j]),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
def create_dataset():
    """
    This functions creates a pandas DataFrame for receiving performance analysis
    
    Returns:
        An empty pandas DataFrame object with classification metrics to be evaluated
    """
    attributes = ['acc', 'prec', 'rec', 'f1', 'auc', 'total_time']
    model_performance = pd.DataFrame({})
    for col in attributes:
        model_performance[col] = []
        
    return model_performance

def model_analysis(model, X, y, X_test, y_test, df_performance, cv=5, train=True):
    """
    This function brings up a full model evaluation and saves it in a DataFrame object
    
    Input:
        model: a classification model already trained
        X, y: data to be evaluated and the target class
        df_performance: an empty dataframe (generated by crate_dataset() function)
        cv: cross validation k folds
        train: a boolean flag for telling it it's a train cv evaluation or a test evaluation
    Returns:
        A DataFrame object with all metrics from all models on classifier's dictionary.
    """
    # Accuracy, precision, recall and f1_score on training set using cv
    t0_cv = time.time()
    acc = cross_val_score(model, X, y, cv=cv, scoring='accuracy').mean()
    prec = cross_val_score(model, X, y, cv=cv, scoring='precision').mean()
    rec = cross_val_score(model, X, y, cv=cv, scoring='recall').mean()
    f1 = cross_val_score(model, X, y, cv=cv, scoring='f1').mean()
    try:
        y_scores = cross_val_predict(model, X, y, cv=5, 
                                 method='decision_function')
    except:
        # Trees don't have decision_function but predict_proba
        y_probas = cross_val_predict(model, X, y, cv=5, 
                                     method='predict_proba')
        y_scores_tree = y_probas[:, 1]
        y_scores = y_scores_tree
    auc = roc_auc_score(y, y_scores)
    # Time spent on cross_validation prediction
    t1_cv = time.time()
    delta_time_cv = t1_cv-t0_cv
    
    # Evaluation using the test set
    t0_test = time.time()
    y_pred_test = model.predict(X_test)
    acc_test = accuracy_score(y_test, y_pred_test)
    prec_test = precision_score(y_test, y_pred_test)
    rec_test = recall_score(y_test, y_pred_test)
    f1_test = f1_score(y_test, y_pred_test)
    y_scores_test = model.predict_proba(X_test)[:, 1]
    auc_test = roc_auc_score(y_test, y_scores_test)
    # Time spent on test prediction
    t1_test = time.time()
    delta_time_test = t1_test-t0_test

    # Saving on dataframe
    performances = {}
    performances['acc'] = round(acc, 4)
    performances['prec'] = round(prec, 4)
    performances['rec'] = round(rec, 4)
    performances['f1'] = round(f1, 4)
    performances['auc'] = round(auc, 4)
    performances['total_time'] = round(delta_time_cv, 3)        
    df_performance = df_performance.append(performances, ignore_index=True)
    
    test_performances = {}
    test_performances['acc'] = round(acc_test, 4)
    test_performances['prec'] = round(prec_test, 4)
    test_performances['rec'] = round(rec_test, 4)
    test_performances['f1'] = round(f1_test, 4)
    test_performances['auc'] = round(auc_test, 4)
    test_performances['total_time'] = round(delta_time_test, 3)        
    df_performance = df_performance.append(test_performances, ignore_index=True)
    
    model_name = model.__class__.__name__
    df_performance.index = [model_name+' cv', model_name+' test']
    
    return df_performance
    
def plot_roc_curve(fpr, tpr, y, y_scores, auc, label=None):
    """
    This function plots the ROC curve of a model
    
    Input:
        fpr: False Positive Rate
        tpr: True Positve Rate
        y: target label
        y_scores: prediction probability scores
        auc: AUC score
        
    Output:
        ROC Curve for the model
    """   
    sns.set(style='white', palette='muted', color_codes=True)
    plt.plot(fpr, tpr, linewidth=2, label=f'{label}, auc={auc:.3f}')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([-0.02, 1.02, -0.02, 1.02])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve', size=14)
    plt.annotate('Minimum ROC Score of 50% \n (This is the minimum score to get)', 
                 xy=(0.5, 0.5), xytext=(0.6, 0.4),
                 arrowprops=dict(facecolor='#6E726D', shrink=0.05))
    plt.legend()
    
def plot_learning_curve(model, ax, X, y, ylim=None, cv=5, n_jobs=1, 
                        train_sizes=np.linspace(.1, 1.0, 10)):
    """
    This functions plots the learning curve for a model using training and cross validation data
    
    Input:
        model: a classification model already trained
        ax: matplotlib axis
        X: training set
        y: training label
        y_lim: y axis limit
        cv: cross validation k folds
        train_sizes: bucket or bins to split the learning curve
    Output:
        Learning curves for models in the trained_models dictionary
    """
    # Returning metrics with sklearn's learning_curve function
    train_sizes, train_scores, test_scores = learning_curve(model, X, y, cv=cv, n_jobs=n_jobs,
                                                           train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    # Ploting the learning curve and standard deviation bin
    ax.fill_between(train_sizes, (train_scores_mean-train_scores_std), (train_scores_mean+train_scores_std),
                    alpha=0.1, color='blue')
    ax.fill_between(train_sizes, (test_scores_mean-test_scores_std), (test_scores_mean+test_scores_std),
                    alpha=0.1, color='crimson')
    ax.plot(train_sizes, train_scores_mean, 'o-', color='navy', label='Training score')
    ax.plot(train_sizes, test_scores_mean, 'o-', color='red', label='Cross-Validation score')
    
    # Plot configuration
    ax.set_title(f'{model.__class__.__name__} Learning Curve', size=14)
    ax.set_xlabel('Training size (m)')
    ax.set_ylabel('Score')
    ax.grid(True)
    ax.legend(loc='best')
    
def plot_precision_vs_recall(precisions, recalls, label=None, color='b'):
    """
    This function plots precision versus recall curve.
    
    Input:
        precisions, recalls: generated by sklearns's precision_recall_curve()
    """
    sns.set(style='white', palette='muted', color_codes=True)
    plt.plot(recalls, precisions, color=color, linewidth=2, label=label)
    plt.title('Precision versus Recall', fontsize=14)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.axis([0, 1, 0, 1])
    plt.legend()
    
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    """
    This function plots precision x recall among different thresholds
    """
    plt.plot(thresholds, precisions[:-1], 'b--', label='Precision')
    plt.plot(thresholds, recalls[:-1], 'g-', label='Recall')
    plt.xlabel('Threshold')
    plt.title('Precision versus Recall - Thresholds', size=14)
    plt.legend(loc='best')
    plt.ylim([0, 1])
```

\begin{equation*}
\text{Accuracy}=\frac{TP+TN}{TP+TN+FP+FN}
\end{equation*}

\begin{equation*}
\text{Precision}=\frac{TP}{TP+FP}
\end{equation*}

\begin{equation*}
\text{Recall}=\frac{TP}{TP+FN}
\end{equation*}

\begin{equation*}
\text{F1-Score}=2\times\frac{\text{Precision} * \text{Recall}}{\text{Precision} + \text{Recall}}
\end{equation*}

# Oversampling

One common way to tackle the issue of imbalanced data is over-sampling. Over-sampling refers to various methods that aim to increase the number of instances from the underrepresented class in the data set.

**Important:** Always split into test and train sets BEFORE trying oversampling techniques! Oversampling before splitting the data can allow the exact same observations to be present in both the test and train sets. This can allow our model to simply memorize specific data points and cause overfitting and poor generalization to the test data.

This **SMOTE** method generates synthetic data based on the feature space similarities between existing minority instances. In order to create a synthetic instance, it finds the K-nearest neighbors of each minority instance, randomly selects one of them, and then calculate linear interpolations to produce a new minority instance in the neighborhood.

```python
# Oversampling
sm = SMOTE(sampling_strategy='minority', random_state=42)
X_train_over, y_train_over = sm.fit_sample(X_train, y_train)

# Plotting results
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(17, 6))
donut_plot('Class', df=df, text=f'Total: \n\n{str(len(df))} samples', ax=axs[0])
axs[0].set_title('Fraud Original \nDataset Proportion', size=14)
donut_plot('Class', df=df_undersampled, text=f'Total: \n\n{str(len(df_undersampled))} samples', ax=axs[1])
axs[1].set_title('Fraud Undersampled \nDataset Proportion', size=14)
df_class_over = pd.DataFrame(y_train_over, columns=['Class'])
donut_plot('Class', df=df_class_over, text=f'Total: \n{str(len(df_class_over))} samples', ax=axs[2])
axs[2].set_title('Fraud Oversampled \nDataset Proportion', size=14)
plt.show()
```

## Threshold Selection

```python
# Dataframe for analysis
threshold_metrics = create_dataset()
threshold_metrics.drop(['total_time'], inplace=True, axis=1)

thresholds = np.linspace(0.90, 1.0, 50)
test_probas = log_reg_over.predict_proba(X_test)[:, 1]
for t in thresholds:
    y_best_recall = np.where(test_probas > t, 1, 0)
    acc = accuracy_score(y_test, y_best_recall)
    recall = recall_score(y_test, y_best_recall)
    precision = precision_score(y_test, y_best_recall)
    f1 = f1_score(y_test, y_best_recall)
    auc = roc_auc_score(y_test, y_best_recall)
    
    # Saving data
    performances = {}
    performances['acc'] = round(acc, 4)
    performances['prec'] = round(precision, 4)
    performances['rec'] = round(recall, 4)
    performances['f1'] = round(f1, 4)
    performances['auc'] = round(auc, 4)
    
    threshold_metrics = threshold_metrics.append(performances, ignore_index=True)
```

```python
# Returning scores 
threshold_index = [round(t, 4) for t in thresholds]
threshold_metrics.index = threshold_index
cm = sns.light_palette("cornflowerblue", as_cmap=True)
threshold_metrics.style.background_gradient(cmap=cm)
```

Now we can ask the business area: "What do you want for Recall, Precision, F1-Score or AUC? Here you have all the tradeoffs"

Let's select 9 possible good thresholds and see the results through confusion matrix

```python
cf_mx_thresholds = threshold_index[-10:-1]
i = 1
plt.figure(figsize=(14, 10))
for t in cf_mx_thresholds:
    custom_scores = np.where(test_probas > t, 1, 0)
    cf_mx = confusion_matrix(y_test, custom_scores)
    plt.subplot(3, 3, i)
    plot_confusion_matrix(cf_mx, classes, title=f'Threshold={t}',
                          cmap=plt.cm.Greens)
    i += 1
plt.tight_layout()
plt.suptitle('Threshold Comparison')
plt.subplots_adjust(top=0.93)
plt.show()
```

# Calibration Curves

When performing classification one often wants to predict not only the class label, but also the associated probability. This probability gives some kind of confidence on the prediction. Well calibrated classifiers are probabilistic classifiers for which the output of the predict_proba method can be directly interpreted as a confidence level. For instance a well calibrated (binary) classifier should classify the samples such that among the samples to which it gave a predict_proba value close to 0.8, approx. 80% actually belong to the positive class.

```python
probs = log_reg_over.predict_proba(X_test)[:, 1]
frac_of_positives, mean_pred_value = calibration_curve(y_test, probs, n_bins=10)

fig = plt.subplots(figsize=(10, 10))
ax1 = plt.subplot2grid((3, 1), (0, 0), rowspan=2)
ax2 = plt.subplot2grid((3, 1), (2, 0))

ax1.plot([0, 1], [0, 1], "k:", label="Perfectly calibrated")
ax1.plot(mean_pred_value, frac_of_positives, "s-", label='Logistic Regression')
ax1.legend(loc="upper left")
ax1.set_title('Logistic Regression Calibration Curve', size=14)
ax1.set_ylabel("Fraction of positives")
ax1.set_ylim([-0.05, 1.05])
format_spines(ax1, right_border=False)

ax2.hist(probs, range=(0, 1), bins=10, histtype="step", lw=2)
ax2.set_xlabel("Mean predicted value")
ax2.set_ylabel("Count")
format_spines(ax2, right_border=True)

plt.tight_layout()
plt.show()
```
