# Lendo e Entendendo os Dados

Para este conjunto de dados em específico, é necessário aumentar o tamanho limite dos campos de leitura de arquivos. Fonte: [thread-stack-overflow](https://stackoverflow.com/questions/15063936/csv-error-field-larger-than-field-limit-131072)

Um outro problema enfrentado para leitura do arquivo foi um erro de permissão negada (Ubuntu). Para resolver este problema, foi preciso executar o comando `sudo chmod 777 <file.csv>`. Fonte: [kaggle-topic-access-denied](https://www.kaggle.com/c/ga-customer-revenue-prediction/discussion/71536)

```python
# Realizando procedimento para permitir leitura dos arquivos
import sys
import csv
maxInt = sys.maxsize

while True:
    # decrease the maxInt value by factor 10 
    # as long as the OverflowError occurs.
    try:
        csv.field_size_limit(maxInt)
        break
    except OverflowError:
        maxInt = int(maxInt/10)
```

```python
# Importando pandas e lendo dados
import pandas as pd

df = pd.read_csv('zomato.csv', sep=',', engine='python')
df.head()
```

```python
# Verificando dimensões e atributos
print(f'O dataset possui {df.shape[0]} linhas e {df.shape[1]} colunas.')

print(f'\nColunas: \n{list(df.columns)}')
```

Vejamos os detalhes sobre cada uma dessas colunas.

* **url:** url do restaurante no website zomato (fornecedor do dataset);
* **address:** endereço do restaurante na cidade de Bengaluru;
* **name:** nome do restaurante;
* **online_order:** flag que informa se o restaurante possibilita, ou não, pedidos online;
* **book_table:** flag que informa se o restaurante possibilita, ou não, agendamento de mesas;
* **rate:** nota geral do restaurante (de 0 a 5);
* **votes:** quantidade de votos para o restaurante até a data do conjunto de dados;
* **phone:** telefone do restaurante;
* **location:** região na qual o restaurante está localizado;
* **rest_type:** tipo do restaurante;
* **dish_liked:** pratos mais adorados pelos clientes do restaurante;
* **cuisines:** estilos de pratos, separados por vírgula;
* **approx_cost(for two people):** custo aproximado para duas pessoas;
* **reviews_list:** lista de tuplas contendo análises do restaurante. Cada tupla contém dois valores: rating e review por cliente;
* **menu_item:** lista de menus disponíveis no restaurante;
* **listed_in(type):** categorização do restaurante quanto ao tipo de comida;
* **listed_in(city):** categorização do restaurante quanto a cidade;

Temos uma grande quantidade de dados nulos que precisarão ser tratadas futuramente antes de entrarmos na parte de análise dos dados. Por enquanto, vamos manter em mente essa quantidade e utilizar preparações intermediárias, se necessário, nessa sessão de transformação dos dados.

## Transformando Tipos Primitivos

```python
# Tipos primitivos
df.dtypes
```

Vamos aplicar algumas transformações nos dados:

* **Valor pago para duas pessoas:**
    - Transformação do tipo primitivo de string para float;
    

* **Nota:**
    - Sabemos que a nota vai de 0 a 5 e, portanto, vamos eliminar o texto "/5" de cada instância e transformar os dados em float

```python
# Transformando em string para substituir vírgula por ponto
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(str)
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].apply(lambda x: x.replace(',', '.'))

# Transformando em float
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].astype(float)
```

```python
# Verificando
print(f'{type(df["approx_cost(for two people)"][0])}')
```

```python
# Transformando dados de ratings
import numpy as np
df['rate_transformed'] = df['rate'].astype(str)
df['rate_transformed'] = df['rate_transformed'].apply(lambda x: x.split('/')[0])

# Cuidando de entradas inválidas
df['rate_transformed'] = df['rate_transformed'].apply(lambda x: x.replace('NEW', str(np.nan)))
df['rate_transformed'] = df['rate_transformed'].apply(lambda x: x.replace('-', str(np.nan)))

# Transformando em float
df['rate_transformed'] = df['rate_transformed'].astype(float)
df.drop(['rate'], axis=1, inplace=True)
```

Vamos validar as alterações realizadas

```python
# Novos tipos primitivos
df.dtypes
```

## Tratando Dados Nulos

Como já visualizamos acima, nosso conjunto de dados possui diversos dados nulos. Em uma primeira abordagem, visando uma rápida análise dos dados, vamos simplesmente dropar as colunas com dados nulos e procurar por insights mais assertivos sobre esse comportamento.

Aliás, seria até mais adequado reservar os dados nulos da coluna `rate` em um conjunto de dados separados e, posteriormente, treinar um modelo para previsão da nota de cada restaurante de acordo com seus demais atributos (a serem definidos, filtrados e selecionados). Antes de realizar qualquer eliminação de linhas, vamos primeiro definir as principais colunas com uma maior possível influência no rate do restaurante.

```python
# Reservando dados nulos
df_unrated = df[df['rate_transformed'].isnull()]

# Eliminando rates nulos do dataset original
df.dropna(subset=['rate_transformed'], inplace=True)

# Verificando
df.isnull().sum()
```

```python
# Eliminando colunas para uma primeira análise
df.drop(['url', 'phone'], axis=1, inplace=True)
```

## Analisando Dados

```python
# Importando bibliotecas úteis
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Top Restaurantes por Rating

Nesta análise, vamos agrupar os dados para visualizar os TOP 10 restaurantes por rating.

```python
# Agrupando dados
grouped_rest = df.groupby(by='name', as_index=False).mean()
top_rating = grouped_rest.sort_values(by='rate_transformed', ascending=False).iloc[:10, np.r_[0, -1]]
top_rating
```

```python
# Ajustando nome do restaurante alterado
top_rating.iloc[1, 0] = 'Spa Cuisine'
```

```python
def format_spines(ax, right_border=True):
    """
    this function sets up borders from an axis and personalize colors
    """    
    # Setting up colors
    ax.spines['bottom'].set_color('#CCCCCC')
    ax.spines['left'].set_color('#CCCCCC')
    ax.spines['top'].set_visible(False)
    if right_border:
        ax.spines['right'].set_color('#CCCCCC')
    else:
        ax.spines['right'].set_color('#FFFFFF')
    ax.patch.set_facecolor('#FFFFFF')
```

```python
# Plotando graficamente
fig, ax = plt.subplots(figsize=(8, 5))
ax = sns.barplot(y='name', x='rate_transformed', data=top_rating, palette='Blues_d')
ax.set_xlim([4.7, 5])
format_spines(ax, right_border=False)

for p in ax.patches:
    width = p.get_width()
    ax.text(width+0.01, p.get_y() + p.get_height() / 2. + 0.2, '{:1.2f}'.format(width), 
            ha="center", color='grey')

ax.set_title('Top 10 Restaurantes em Bengaluru', size=16)
plt.show()
```

### Top Restaurantes por Custo

Vejamos algumas estatísticas sobre o custo aproximado (para duas pessoas) de cada restaurante.

```python
# Estatísticas
df['approx_cost(for two people)'].describe()
```

Temos restaurante(s) onde o custo aproximado para duas pessoas é de **1** unidade monetária indiana!

```python
# Retornando dados
high_cost = grouped_rest.sort_values(by='approx_cost(for two people)', 
                                     ascending=False).iloc[:10, np.r_[0, -1, -2]]

# Plotando mais custosos
fig, ax = plt.subplots(figsize=(10, 7))
sns.barplot(x='name', y='approx_cost(for two people)', data=high_cost, ax=ax, palette='PuBu')
ax2 = ax.twinx()
sns.lineplot(x='name', y='rate_transformed', data=high_cost, ax=ax2, color='crimson', sort=False)
ax.tick_params(axis='x', labelrotation=90)
ax.set_ylim(700, 980)
ax2.set_ylim([3, 5])
format_spines(ax, right_border=True)
format_spines(ax2, right_border=True)
ax.xaxis.set_label_text("")

xs = np.arange(0,10,1)
ys = high_cost['rate_transformed']

for x,y in zip(xs,ys):
    label = "{:.2f}".format(y)
    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center', # horizontal alignment can be left, right or center
                 color='black')

ax.set_title('Restaurantes Com Maior Custo Médio e suas Notas', size=14)
plt.tight_layout()
```

```python
# Retornando dados
low_cost = grouped_rest.sort_values(by='approx_cost(for two people)', 
                                     ascending=True).iloc[:10, np.r_[0, -1, -2]]

# Plotando mais custosos
fig, ax = plt.subplots(figsize=(10, 7))
sns.barplot(x='name', y='approx_cost(for two people)', data=low_cost, ax=ax, palette='PuBu')
ax2 = ax.twinx()
sns.lineplot(x='name', y='rate_transformed', data=low_cost, ax=ax2, color='crimson', sort=False)
ax.tick_params(axis='x', labelrotation=90)
ax.set_ylim([0, 2])
ax2.set_ylim([0, 5])
format_spines(ax, right_border=True)
format_spines(ax2, right_border=True)
ax.xaxis.set_label_text("")

xs = np.arange(0,10,1)
ys = low_cost['rate_transformed']

for x,y in zip(xs,ys):
    label = "{:.2f}".format(y)
    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center', # horizontal alignment can be left, right or center
                 color='black')

ax.set_title('Restaurantes Com Menor Custo Médio e suas Notas', size=14)
plt.tight_layout()
```

**Conclusão:**

O custo médio (para duas pessoas) não é um item mandatório na questão de rating do restaurante. Existem restaurantes caros com notas baixas e também existem restaurantes muito baratos notas altas.

### Relação Entre Custo e Nota

Para exemplificar de forma mais técnica a afirmação acima, vamos plotar a distribuição entre o custo aproximado para duas pessoas e o rating de cada um dos restaurantes.

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(x='rate_transformed', y='approx_cost(for two people)', data=df, ax=ax)
format_spines(ax, right_border=False)
ax.set_title('Relação Entre Nota e Custo Aproximado', size=14)
plt.show()
```

```python
# Visualizando distribuição por demais atributos
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))
sns.scatterplot(x='rate_transformed', y='approx_cost(for two people)', hue='online_order', 
                data=df, ax=axs[0])
sns.scatterplot(x='rate_transformed', y='approx_cost(for two people)', hue='book_table', 
                data=df, ax=axs[1])
format_spines(axs[0], right_border=False)
format_spines(axs[1], right_border=False)
axs[0].set_title('Distribuição por Delivery', size=14)
axs[1].set_title('Distribuição por Reserva de Mesa', size=14)
plt.show()
```

**Conclusões:**

* Realmente a nota de cada restaurante não tem correlação direta com seu custo médio para duas pessoas e o gráfico de distribuição pode provar essa afirmação;
* A segunda plotagem que relaciona a possibilidade de entrega e reserva de mesa mostra que:
    * Existe uma leve concentração de restaurantes que não possuem serviço de entrega e com um baixo custo médio por pessoa. Analogamente, uma leve distribuição de restaurantes com serviço de entrega encontra-se na parte de cima do eixo do custo. Provavelmente, a falta desse serviço atua na diminuição do custo de cada restaurante.
    * Pouquíssimos restaurantes oferecem serviço de reserva de mesa. Aqueles que oferecem, geralmente possuem as maiores notas.

### Serviços de Entrega e Reserva

Apenas para comprovar a relação de contagem acima proposta, vamos visualizar a quantidade de restaurantes que possuem os serviços de entrega e de reserva de mesa.

```python
df.head(1)
```

```python
# Contagem de restaurantes por oferta de Delivery
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

sns.countplot(x='online_order', data=df, ax=axs[0], palette='Blues_d')
format_spines(axs[0], right_border=False)
ncount = len(df)
for p in axs[0].patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        axs[0].annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), 
                ha='center', va='bottom') # set the alignment of the text
axs[0].set_title('Contagem de Restaurantes por Oferta de Delivery', size=14)

# Contagem de restaurantes por agendamento de mesa
sns.countplot(x='book_table', data=df, ax=axs[1], palette='Blues_d')
format_spines(axs[1], right_border=False)
ncount = len(df)
for p in axs[1].patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        axs[1].annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), 
                ha='center', va='bottom') # set the alignment of the text
axs[1].set_title('Contagem de Restaurantes por Agendamento de Mesa', size=14)

plt.tight_layout()
plt.show()
```

Vamos visualizar o preço médio para duas pessoas e o rating médio para cada um dos casos

```python
# Comparação de restaurantes por oferta de delivery
df_delivery = df.groupby(by='online_order').mean()
df_delivery
```

```python
# Comparação de restaurantes por agendamento de mesa
df_delivery = df.groupby(by='book_table').mean()
df_delivery
```

**Conclusões:**

* Restaurantes que ofertam agendamento de mesa possuem uma maior média de rating em comparação aos restaurantes que não ofertam estse serviço. Curiosamente, o custo médio dos restaurantes com agendamento é significamente menor.
* Estabalecimentos com oferta de delivery possuem um custo médio maior em relação aos restaurantes sem a possibilidade de entrega. Com relação as notas, o valor é próximo um do outro.
* Clientes ficam mais satisfeitos com restaurantes que possuem agendamento de mesa.

### Categorias de Restaurante

Vamos analisar os custos e ratings a partir do tipo do restaurante.

```python
# Visualizando entradas possíveis
df['listed_in(type)'].value_counts()
```

```python
# Atributos médios por tipo
type_rest = df.groupby(by='listed_in(type)').mean().sort_values(by='rate_transformed', ascending=False)
type_rest
```

```python
# Retornando dados
rest_params = df.groupby(by='listed_in(type)', as_index=False).mean().sort_values(by='rate_transformed', 
                                                                                  ascending=False)

# Plotando mais custosos
fig, ax = plt.subplots(figsize=(10, 7))
sns.barplot(x='listed_in(type)', y='approx_cost(for two people)', data=rest_params, ax=ax, palette='Blues_d')
ax2 = ax.twinx()
sns.lineplot(x='listed_in(type)', y='rate_transformed', data=rest_params, ax=ax2, color='crimson', sort=False)
ax.tick_params(axis='x', labelrotation=90)
"""ax.set_ylim([0, 2])
ax2.set_ylim([0, 5])"""
format_spines(ax, right_border=True)
format_spines(ax2, right_border=True)
ax.xaxis.set_label_text("")

xs = np.arange(0,10,1)
ys = rest_params['rate_transformed']

for x,y in zip(xs,ys):
    label = "{:.2f}".format(y)
    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center', # horizontal alignment can be left, right or center
                 color='black')

ax.set_title('Custo Médio e Rating de Restaurantes por Categoria', size=14)
plt.tight_layout()
```

**Conclusões:**

* Restaurantes do tipo _Pubs e Bares_ e _Drinks e Casas Noturnas_ possuem os dois menores custos médios (para duas pessoas) e as duas maiores notas. São os preferidos pelos clientes provavalmente pelo baixo custo oferecido.

* Restaurantes de _Delivery_ e aqueles específicos para _Jantares_ possuem as menores notas entre os clientes. É válido dizer que esse tipo de restaurante possui uma cliente-la extremamente exigente.

* Restaurantes do tipo _Café_ são os mais caros.

### Localidade de Restaurantes

Vamos repetir a análise e observar custos e notas dos restaurantes a partir de sua localidade (cidade)

```python
# Entradas possíveis
df['listed_in(city)'].value_counts()
```

```python
# Atributos médios por tipo
city_rest = df.groupby(by='listed_in(city)').mean().sort_values(by='rate_transformed', ascending=False)
city_rest
```

```python
# Retornando dados
city_rest = df.groupby(by='listed_in(city)', as_index=False).mean().sort_values(by='rate_transformed', 
                                                                                  ascending=False)

# Plotando mais custosos
fig, ax = plt.subplots(figsize=(14, 7))
sns.barplot(x='listed_in(city)', y='approx_cost(for two people)', data=city_rest, ax=ax, palette='Blues_d')
ax2 = ax.twinx()
sns.lineplot(x='listed_in(city)', y='rate_transformed', data=city_rest, ax=ax2, color='crimson', sort=False)
ax.tick_params(axis='x', labelrotation=90)
format_spines(ax, right_border=True)
format_spines(ax2, right_border=True)
ax.xaxis.set_label_text("")

xs = np.arange(0,len(city_rest),1)
ys = city_rest['rate_transformed']

for x,y in zip(xs,ys):
    label = "{:.2f}".format(y)
    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center', # horizontal alignment can be left, right or center
                 color='black')

ax.set_title('Custo Médio e Rating de Restaurantes por Cidade', size=14)
plt.tight_layout()
```

**Conclusões:**

* Restaurantes localizados em Church Street, MG Road e Brigade Road são os preferidos entre os clientes. Seus custos são baixos e suas notas são altas.
* Em contrapartida, restaurantes localizados em Electronic City são aqueles que possuem o menor rating. Ademais, seu custo pode ser considerado mediano.
* Os restaurantes com maior custo médio para duas pessoas estão localizados em Kaiyan Nagar.
